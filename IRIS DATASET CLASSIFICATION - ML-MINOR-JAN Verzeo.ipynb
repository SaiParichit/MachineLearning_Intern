**Create a Set of Independent Variables and Dependent Variable for Training and Testing Models.**

```python
X = Iris.drop("Species", axis = "columns")
Y = Iris["Species"]
```

**Import Library for Splitting the Data Set into Test and Test Data Sets**

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
print(f"80 % of Train Data Set Size : {len(X_Train)} and Data Set Size of Training Labels : {len(Y_Train)}")
print()
print(f"20 % of Test Data Set Size : {len(X_Test)} and Data Set Size of Testing Labels : {len(Y_Test)}")
```

**Import Libraries for Model Selection and Algorithms.**

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

```python
from sklearn.linear_model import LogisticRegression

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.neighbors import KNeighborsClassifier

from sklearn.tree import DecisionTreeClassifier

from sklearn.naive_bayes import GaussianNB

from sklearn.svm import SVC
```

**Tuples of Machine Learning Models and their Parameters Appended to form a List of Models.**

```python
Models = []
Models.append(("LR", LogisticRegression(solver = "liblinear", multi_class = "ovr")))
Models.append(("LDA", LinearDiscriminantAnalysis()))
Models.append(("KNN", KNeighborsClassifier()))
Models.append(("CART", DecisionTreeClassifier()))
Models.append(("GNB", GaussianNB()))
Models.append(("SVM", SVC(gamma = "auto")))
```

**Import Libraries for Model Selection**

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

### Accuracy of Models.

```python
Results = []
Names = []

for Name, Model in Models :
    KFold = StratifiedKFold(n_splits = 10, random_state = 1, shuffle = True)
    CrossValidation = cross_val_score(Model, X_Train, Y_Train, cv = KFold, scoring = "accuracy")
    Results.append(CrossValidation)
    Names.append(Name)
    print(f"{Name} Accuracy : {CrossValidation.mean()*100:.2f}%")
```

**n_splits :** Number of Folds.

**shuffle :** Shuffles Each Sample before Splitting.

**random_state :** Seed used by Random Number Generator to make Results Reproducible.

**cv :** Cross Validation ( An Iterable ) Number of Iterations we have to Set.

**scorer :** Evaluation Metric we want to Calaculate for the Accuracy Score of the Model.

### Conclusion : SVM is giving the Best Accuracy.
****
**Import Library for Data Visualization**

```python
import matplotlib.pyplot as plt
```

**Compare Algorithms by Visualization for Better Understanding.**

```python
plt.figure(figsize = (12,10))
plt.boxplot(Results, labels = Names)
plt.title("Algorithm Comparison");
```

### Prediction on Test Data Set.

**Support Vector Classifier is the Most Accurate Model for the Data Set**

```python
Model = SVC(gamma = "auto")
```

**Fitting the Model on Train and Test Data Set to Find the Patterns in the Data.**

```python
Model.fit(X_Train, Y_Train)
```

**Prediction on Test Data Set**

```python
Y_Prediction = Model.predict(X_Test)
```

### Evaluation of Predictions made by the Model.

**Import Libraries for Evaluation of Predictions made by Support Vector Classifier**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

**Evaluate Predictions**

```python
print(f"Accuracy Score of Support Vector Classifier : {accuracy_score(Y_Test, Y_Prediction)*100:.2f}%")
```

**Confusion Matrix**

```python
print(f"Confusion Matrix for Support Vector Classifier : \n{confusion_matrix(Y_Test, Y_Prediction)}")
```

**Classification Report of Support Vector Classifier**

```python
print(f"Classificaton Report of Support Vector Classifier : \n{classification_report(Y_Test, Y_Prediction)}")
```
